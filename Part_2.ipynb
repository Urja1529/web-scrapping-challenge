```python
#Import Splinter and BeautifulSoup
from splinter import Browser
from bs4 import BeautifulSoup 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
browser = Browser('chrome')
```

```python
url = 'https://static.bc-edx.com/data/web/mars_facts/temperature.html'
browser.visit(url)
```

```python
html=browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
entire_table=soup.find_all('div',class_='container-fluid py-5')
```

```python
table_data=[]
for i in entire_table:
    j=i.find_all('tr')
    for item in j:
        row=item.text
        row=row.replace('\n',' ')
        table_data.append(row)
        
table_data
```

```python
table_data = pd.read_html("https://static.bc-edx.com/data/web/mars_facts/temperature.html")
table_data
```

```python
# df=pd.DataFrame(table_data)
df=table_data[0]
df
```

```python
df.dtypes
```

```python
df['terrestrial_date']= pd.to_datetime(df['terrestrial_date'])
```

```python
df.dtypes
```

```python
# How many months exist on Mars?
month_count=df['terrestrial_date'].dt.month.count()
month_count
```

```python
# df['terrestrial_date'].groupby(df.terrestrial_date.dt.month).agg('count')
```

```python
df.groupby(df.terrestrial_date.dt.month)['terrestrial_date'].count()
```

```python
month_count=df['month'].value_counts()
# mpnth_count=month_count.sort_values()
month_count
```

```python
# How many Martian (and not Earth) days worth of data exist in the scraped dataset?
df['sol'].count()
```

```python
# Find the average minimum daily temperature for all of the months.
avg_min=df.groupby(by=['month']).mean()["min_temp"]
avg_min
```

```python
# Plot the results as a bar chart.
bar_chart=avg_min.plot.bar()
bar_chart
```

```python
coldest_df=df.groupby(by=['month']).mean()["min_temp"].sort_values()
coldest_df
```

```python
bar_chart=coldest_df.plot.bar()
bar_chart
```

```python
# Which months have the lowest and the highest atmospheric pressure on Mars? To answer this question:
# Find the average daily atmospheric pressure of all the months.
avg_pressure=df.groupby(by=['month']).mean()['pressure'].sort_values()
avg_pressure
```

```python
# Plot the results as a bar chart.
pressure_chart=avg_pressure.plot.bar()
pressure_chart
```
